```python

kamus_alay = {
    'tpi': 'tapi',
    'd': 'di',
    'ak': 'aku',
    'kl': 'kalau',
    'grgr': 'garagara',
    'x': 'kali',
    'bgt': 'banget',
    'bangt': 'banget',
    'blng': 'bilang',
    'w': 'aku',  # atau 'gue' sesuai preferensi
    'gw': 'gue',
    'gk': 'gak',
    'ga': 'nggak',
    'sm': 'sama',
    'tp': 'tapi',
    'jg': 'juga',
    'udh': 'sudah',
    'dr': 'dari',
    'krn': 'karena',
    'sy': 'saya',
    'trs': 'terus',
    'lg': 'lagi',
    'aja': 'saja',
    'pdhl': 'padahal',
    'pls': 'please',
    'blajar': 'belajar',
    'bljr': 'belajar',
    'jd':'jadi',
    'jdi':'jadi',
    'jg':'juga',
    'org':'orang',
    'u': 'kamu',
    'rl':'real life',
    'trnyt':'terus',
    'sampe':'sampai',
    'pny':"punya",
    'aq':'aku',
    'adl':'adalah',
    'knp':'kenapa',
    'bbrp':'beberapa',
    'bsk':'besok',
    'menikaj':'menikah',
    'mg':'memang',
    'kl':'kalau',
    'klo':'kalau',
    'blm':'belum',
    'tep':'tetap',
    'kyk':'seperti',
    'dri':'dari',
    'dr':'dari',
    'gtu':'begitu',
    'skrg':'sekarang'
    # tambahkan kata lain sesuai kebutuhan
}

def normalisasi_alay(text):
    # Ganti kata alay sesuai kamus
    words = text.split()
    words = [kamus_alay.get(word, word) for word in words]
    text = ' '.join(words)
    return text

def hapus_double_kata(text):
    # Gabungkan double kata (misal: 'jadi jadi' -> 'jadi')
    text = re.sub(r'\b(\w+)( \1\b)+', r'\1', text)
    return text

def hapus_double_huruf(text):
    # Gabungkan double huruf (misal: 'bangettt' -> 'banget')
    text = re.sub(r'(.)\1{2,}', r'\1', text)  # 3 huruf atau lebih jadi 1
    text = re.sub(r'(.)\1+', r'\1', text)     # double huruf jadi 1
    return text

def bersihkan_alay(text):
    text = text.lower()
    text = normalisasi_alay(text)
    text = hapus_double_kata(text)
    text = hapus_double_huruf(text)
    return text.strip()
```

```python
# Install Sastrawi jika belum
# !pip install Sastrawi

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
# from Sastrawi.StopWordRemover.StopWordRemover import StopWordRemover
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory


# Buat stopword factory dan tambahkan stopword custom
factory = StopWordRemoverFactory()
stopwords = set(factory.get_stop_words())
stemmer = StemmerFactory().create_stemmer()
# Tambahkan stopword custom sesuai kebutuhan
stopwords.update([
    'maksudnya', 'papa', 'kayaknya', 'semua', 'yg', 'kan', 'padahal', 'ketika', 'malah', 'twt', 'tweet', 'wkwk', 'wk', 'ha', 'jir', 'anjir', 'mjb', 'that', 'd', 'spt', 'x', 'aja', 'banget', 'kayak', 'gitu', 'nih', 'sih', 'dong', 'deh', 'loh', 'mah', 'nya', 'nyaa', 'yaa', 'lah', 'kan', 'pun', 'toh', 'kok', 'dong', 'gua', 'gue', 'loe', 'lu', 'loe', 'gw', 'ya', 'ga', 'nggak', 'ngga', 'gak', 'udah', 'dah', 'lagi', 'malah', 'sama', 'buat', 'biar', 'bikin', 'bakal', 'bakalan', 'ntar', 'ntar', 'sampe', 'sampe', 'padahal', 'walaupun', 'walau', 'walaupun', 'meskipun', 'walau', 'walaupun', 'meski', 'meskipun'
])

def bersihkan_dan_stopword(text):
    # Lowercase dan hapus newline/tab
    text = text.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ').lower()
    # Hapus link
    text = re.sub(r'http\S+|www\.\S+|[a-zA-Z0-9.-]+\.(com|id|org|net|co|ac|go|edu|my|ly|me|co.id|web.id|sch.id|xyz|info|io|ai|gov|biz|tv|cc|us|uk|sg|jp|kr|cn|ru|au|fr|de|es|it|nl|se|no|fi|dk|pl|cz|tr|in|za|br|mx|ca|ar|ch|at|be|pt|gr|ro|ua|by|lt|lv|ee|sk|hu|bg|rs|hr|si|ba|mk|al|ge|am|az|kg|kz|tj|tm|uz|vn|th|ph|my|sg|hk|tw|mo|pk|bd|lk|np|mv|ir|iq|il|sa|ae|qa|kw|om|bh|ye|jo|lb|sy|ps|af|mn|kh|la|mm|bt|tl|bn|id|my|sg|th|vn|ph|kr|jp|cn|tw|hk|mo|mn|ru|kz|uz|kg|tj|tm|az|am|ge|tr|cy|gr|bg|ro|md|ua|by|pl|lt|lv|ee|fi|se|no|dk|is|ie|uk|nl|be|lu|fr|de|ch|at|cz|sk|hu|si|hr|ba|rs|me|al|mk|it|sm|va|mt|es|pt|ad|mc|li|gl|fo|ax|sj|eu|int|arpa|mil|edu|gov|biz|info|name|pro|aero|coop|museum|asia|cat|jobs|mobi|tel|travel|xxx|id|co.id|web.id|sch.id|or.id|ac.id|go.id|mil.id|my.id|biz.id|desa.id|ponpes.id|sch.id|web.id|id|co|or|ac|sch|go|mil|my|biz|desa|ponpes|web|sch|id)\b', '', text)
    # Hapus angka dan bullet
    text = re.sub(r'\d+\.', '', text)
    # Hapus karakter non-alfabet (kecuali spasi)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text)
    # Hapus stopword dan stwmming
    text = ' '.join([word for word in text.split() if word not in stopwords])
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text.strip()

df['cleaned'] = df['Content'].astype(str).apply(bersihkan_dan_stopword).apply(bersihkan_alay)
```
