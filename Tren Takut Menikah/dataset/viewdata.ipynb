```python
import pandas as pd
```

```python
data = pd.read_csv('takut-nikah.csv')
data.head(100)  # Display the first 10 rows of the DataFrame
```

```python
# cara ambil kolom hadnle dan content saja
data_filtered = data[['Handle', 'Content', 'Timestamp']]
data_filtered.head(10)  # Display the first 10 rows of the filtered DataFrame
```

```python
# convert timestamp to datetime indonesian
data['Timestamp'] = pd.to_datetime(data['Timestamp'], utc=True).dt.tz_convert('Asia/Jakarta')
data['Timestamp'] = data['Timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')
data_filtered['Timestamp'] = pd.to_datetime(data_filtered['Timestamp'], utc=True).dt.tz_convert('Asia/Jakarta')
data_filtered['Timestamp'] = data_filtered['Timestamp'].dt.strftime('%Y-%m-%d')

data_filtered.head(10)  # Display the first 10 rows of the filtered DataFrame with formatted timestamp
```
